```python
from pathlib import Path
import pandas as pd
import plotly.express as px
```

```python
img_dir = Path('..') / 'report' / 'notebook_01' / 'img'
img_dir.mkdir(parents=True, exist_ok=True)

data_file = Path('..') / 'data' / 'data_processed' / 'SUPERSTORE_TRATADO.csv'
df = pd.read_csv(data_file, sep=',')
```

```python
fig = px.histogram(
    df,
    x='ORDER_DATE',
    y='PROFIT',
    color='CATEGORY',
)
fig.write_image(img_dir / 'histogram_profit_by_category.png')
fig.show()
```

```python
df_map = df.groupby('STATE')['PROFIT'].sum().reset_index()

fig = px.choropleth(
    df_map,
    locations='STATE',
    locationmode='USA-states',  
    color='PROFIT',
    scope='usa',  
    color_continuous_scale='RdYlGn'
)

fig.write_image(img_dir / 'choropleth_profit_sum_by_state.png')
fig.show()
```

```python
df_map = df.groupby('STATE')['PROFIT'].mean().reset_index()

fig = px.choropleth(
    df_map,
    locations='STATE',
    locationmode='USA-states',  
    color='PROFIT',
    scope='usa',  
    color_continuous_scale='RdYlGn'
)

fig.write_image(img_dir / 'choropleth_profit_mean_by_state.png')
fig.show()
```

```python
fig = px.pie(df, values='PROFIT', names='REGION', title='Lucro por Região')
fig.write_image(img_dir / 'pie_profit_by_region.png')
fig.show()
```

```python
cliente = df['SHIP_MODE'].value_counts().reset_index()
cliente.columns = ['SHIP_MODE', 'COUNT']

fig = px.pie(
    cliente,
    values='COUNT',
    names='SHIP_MODE',
    title='SHIP_MODE mais frequente (por número de pedidos)'
)

fig.write_image(img_dir / 'pie_ship_mode_frequency.png')
fig.show()
```
