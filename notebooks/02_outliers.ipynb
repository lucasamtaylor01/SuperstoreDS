```python
import pandas as pd
import plotly.express as px
import numpy as np
from pathlib import Path
```

```python
data_file = Path("..") / "data" / "data_processed" / "SUPERSTORE_DATASET_TRATADO.csv"
df = pd.read_csv(data_file, sep=",")
```

```python
df.head()
```

```python
df_filtered = df.copy()
analise_de_outliers = ['PROFIT', 'NET_SALES']

for col in analise_de_outliers:
    fig_1 = px.histogram(df_filtered, x=col, title=col)
    fig_2 = px.box(df_filtered, y=col, title=col)
    fig_1.show()
    fig_2.show()
```

```python


df_filtered['NET_SALES_LOG'] = np.log1p(df['NET_SALES'])

analise_de_outliers = ['PROFIT', 'NET_SALES_LOG']


for col in analise_de_outliers:
    Q1 = df_filtered[col].quantile(0.15)
    Q3 = df_filtered[col].quantile(0.85)
    IQR = Q3 - Q1
    df_filtered = df_filtered[(df_filtered[col] >= Q1 - 1.5*IQR) & (df_filtered[col] <= Q3 + 1.5*IQR)]

print(round((df.shape[0] - df_filtered.shape[0]) / df.shape[0],4)*100)
```

```python
for col in analise_de_outliers:
    fig_1 = px.histogram(df_filtered, x=col, title=col)
    fig_2 = px.box(df_filtered, y=col, title=col)
    fig_1.show()
    fig_2.show()
```
