```python
# BIBLIOTECAS UTILIZADAS

from pathlib import Path

import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# CARREGANDO O DATAFRAME

data_file = Path('..') / 'data' / 'data_processed' / 'SUPERSTORE_TRATADO.csv'
df_filtered = pd.read_csv(data_file, sep=',')
```

```python
# SELEÇÃO DE COLUNAS PARA ANÁLISE DE OUTLIERS

analise_de_outliers = ['PROFIT', 'NET_SALES']

for col in analise_de_outliers:
    fig_1 = px.histogram(df_filtered, x=col, title=col)
    fig_2 = px.box(df_filtered, y=col, title=col)
    fig_1.show()
    fig_2.show()
```

```python
# TRANSFORMAÇÃO PARA ESCALA LOGARÍTMICA DE NET_SALES

df_filtered['NET_SALES_LOG'] = np.log1p(df_filtered['NET_SALES'])

# TRATAMENTO DE OUTLIERS

analise_de_outliers = ['PROFIT', 'NET_SALES_LOG']
for col in analise_de_outliers:
    Q1 = df_filtered[col].quantile(0.15)
    Q3 = df_filtered[col].quantile(0.85)
    IQR = Q3 - Q1
    df_filtered = df_filtered[(df_filtered[col] >= Q1 - 1.5*IQR) & (df_filtered[col] <= Q3 + 1.5*IQR)]

print(round((df_filtered.shape[0] - df_filtered.shape[0]) / df_filtered.shape[0],4)*100)
```

```python
# VERIFICAÇÃO DE RESULTADOS APÓS O TRATAMENTO DE OUTLIERS

for col in analise_de_outliers:
    fig_1 = px.histogram(df_filtered, x=col, title=col, color_discrete_sequence=['red'])
    fig_2 = px.box(df_filtered, y=col, title=col, color_discrete_sequence=['red'])
    fig_1.show()
    fig_2.show()
```
